```python
from bs4 import BeautifulSoup
import requests
import lxml
import pandas as pd

```

```python
res = requests.get("https://www.flipkart.com/search?q=one%20plus%20phone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
txt = res.text
print(txt)
```

```python
soup = BeautifulSoup(txt,"lxml")
```

```python
# Find all image sources
images = soup.find_all("img")
for img in images:
    print("Image Source:", img.get("src"))

```

```python
price_all = soup.find_all("div", {"class": "_30jeq3 _1_WHN1"})
prices = []
# Find product prices
for price_element in price_all:
    price = price_element.text
    prices.append(price)

device_name=soup.find_all("div",{"class":"_4rR01T"})
devices=[]
for device_element in device_name:
    device = device_element.text
    devices.append(device)

# Now, 'prices' list contains all the product prices
print(prices)


```

```python

data = {"Device": devices, "Prices": prices}
df = pd.DataFrame(data)
```

```python
print(len(devices))
print(len(prices))
print(prices)
```

```python
excel_file = "data.xlsx"
csv_file = "data.csv"
df.to_excel(excel_file, index=False)
df.to_csv(csv_file,index=False)
print(f"Data exported to {excel_file}")
```
